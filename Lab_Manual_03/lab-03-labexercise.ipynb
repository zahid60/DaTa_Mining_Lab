```python
import pandas as pd
import numpy as np
df = pd.read_csv("/kaggle/input/fast-food-calories-data/fastfood_calories.csv")
print(df.head(5))
df.info()
```

```python
missing_fiber = df['fiber'].isnull()
missing_fiber_values = df[missing_fiber]['fiber']

print("Missing 'fiber' values:")
print(missing_fiber_values)
```

```python
# impute null values
updated_df = df
updated_df['fiber']=updated_df['fiber'].fillna(updated_df['fiber'].mean())
updated_df.info()
```

```python
missing_fiber = updated_df['fiber'].isnull()
missing_fiber_values = updated_df[missing_fiber]['fiber']

print("Missing 'fiber' values:")
if not missing_fiber_values.empty:
    print(missing_fiber_values)
else:
    print("Nothing is missing in fiber")
```
