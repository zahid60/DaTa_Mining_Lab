```python
import pandas as pd
import numpy as np
df = pd.read_csv("/kaggle/input/titanic/train_and_test2.csv")
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
updated_df = df
plot = sns.displot(updated_df['Age'], bins=15, kde=True, rug=False)
plt.savefig('/kaggle/working/displot.png')
plt.show()
```

```python
plot=sns.catplot(x="Sex", y="Age", kind="box", data=updated_df)
plt.savefig('/kaggle/working/boxplot.png')
plt.show()
```

```python
plot=sns.violinplot(x="Sex",y="Age",data=updated_df,size=6)
plt.savefig('/kaggle/working/violinplot.png')
plt.show()
```

```python
plot=sns.catplot(x="Sex", y="Age", kind="bar", data = updated_df)
plt.savefig('/kaggle/working/barplot.png')
plt.show()
```

```python
plot=sns.jointplot(x="Sex", y="Age", data=updated_df)
plt.savefig('/kaggle/working/scatterplot.png')
plt.show()
```

```python
plot=sns.catplot(x="Sex", y="Age", hue="Pclass", kind="bar", data = updated_df)
plt.savefig('/kaggle/working/multiplot.png')
plt.show()
```

```python
print(updated_df["Age"].corr(updated_df["Sex"]))
print(updated_df["Age"].corr(updated_df["Sex"], method='spearman'))
# pearson', 'spearman', 'kendall', or a callable, 'zero' 
```

# Perasons correlation visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Creating DataFrames
dt = {'A': np.random.rand(100), 'B': np.random.rand(100)}
dt2 = {'C': np.random.rand(100), 'D': np.random.rand(100)}
data_frame1 = pd.DataFrame(dt)
data_frame2 = pd.DataFrame(dt2)

# Calculating Pearson's correlation
correlation = data_frame1.corrwith(data_frame2)

# Visualize the correlation using a scatter plot
plt.figure(figsize=(6, 6))
plt.scatter(data_frame1['A'], data_frame2['C'], label='A vs C', color='b', marker='o')
plt.scatter(data_frame1['B'], data_frame2['D'], label='B vs D', color='r', marker='x')
plt.title("Scatter Plot of Correlation")
plt.xlabel("DataFrame 1")
plt.ylabel("DataFrame 2")
plt.legend()
plt.grid(True)
plt.show()
```
